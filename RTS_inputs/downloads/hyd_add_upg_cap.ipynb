```python

df_hyd_adj_cap = pd.read_csv('D:\\akash\\reeds\\ReEDS-2.0\\inputs\\supplycurvedata\\hyd_add_upg_cap.csv', header=0)
df_hyd_adj_cap
```

```python
df_hyd_adj_cap['i'] = df_hyd_adj_cap['i'].str.replace('hydUD','hydED')
df_hyd_adj_cap['i'] = df_hyd_adj_cap['i'].str.replace('hydUND','hydEND')
df_hyd_adj_cap
```

```python
#filtering by hydro tech data
grp_tech_nems = df_reeds_nems_genDB.loc[(df_reeds_nems_genDB['tech'] == 'hydED') | (df_reeds_nems_genDB['tech'] == 'hydEND')]
grp_tech_nems


```

```python
df_groupedbyFIPS = grp_tech_nems.groupby(['tech','reeds_ba','reeds_nodal'])['cap'].sum().reset_index()
df_groupedbyFIPS
```

```python
df_groupedbyFIPS.rename(columns = {'cap':'cap_total_bynode'}, inplace = True)
df_groupedbyFIPS
```

```python
df_groupedbyBA = df_groupedbyFIPS.groupby(['tech','reeds_ba'])['cap_total_bynode'].sum().reset_index()
df_groupedbyBA
```

```python
df_groupedbyBA.rename(columns = {'cap_total_bynode':'cap_total_byBA'}, inplace = True)
df_groupedbyBA
```

```python
df_grp_totals = pd.merge(df_groupedbyFIPS, df_groupedbyBA, on=('reeds_ba','tech'), how='inner')
df_grp_totals
```

```python
df_grp_totals['proportion_ofnodeinBA']=df_grp_totals['cap_total_bynode']/df_grp_totals['cap_total_byBA']
df_grp_totals
```

```python
df_hydro_nemsDb_merge = pd.merge(df_grp_totals, df_hyd_adj_cap, left_on=('reeds_ba','tech'), right_on=('r','i'), how='inner')
df_hydro_nemsDb_merge

```

```python
# #df_hydro_nemsDb_merge
# # Write dataframe to csv file (UNCOMPRESSED)
# df_hydro_nemsDb_merge.to_csv('ReEDS2\CSV_County_hydro_CHECK.csv', sep=',', encoding='utf-8', index=False)
```

```python
new_hydro_adj_cap = df_hydro_nemsDb_merge.iloc[:, 9:].multiply(df_hydro_nemsDb_merge['proportion_ofnodeinBA'], axis="index")
new_hydro_adj_cap
```

```python
df_newhydro = pd.merge(df_hydro_nemsDb_merge, new_hydro_adj_cap, left_index=True, right_index=True)
df_newhydro
```

```python
# Dropping columns from df that are not required

df_newhydro.drop(['tech', 'reeds_ba','cap_total_bynode','cap_total_byBA', 'proportion_ofnodeinBA', 'r'], axis=1, inplace=True)
df_newhydro
```

```python
#Renaming FIPS column as per initial file
df_newhydro.rename(columns = {'reeds_nodal':'r'}, inplace = True)
#df_newload_joinedtoorig

#Reindexing or changing datatypes is for ensuring correct joins and operations
df_newhydro = df_newhydro.reindex(axis=1)
#df_newload_joinedtoorig

#Adding the "p" back in fips column
# df_newhydro['r'] = 'p' + df_newhydro['r'].astype(str)
df_newhydro
```

```python
#Droppping duplicate columns with older values of load
df_newhydro.drop(df_newhydro.iloc[:, 3:39], axis=1, inplace=True)

#Cleaning up column names after join caused X & Y additions in names
df_newhydro.columns = df_newhydro.columns.str.replace("_y", "")

#View dataframe and some samples of FIPS length
df_newhydro
```
